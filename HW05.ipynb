```python
import requests
import io
import pandas
```

```python
url = "http://finance.google.com/finance/historical"
```

```python
params = {}
params['q'] = 'TSLA'
params['startdate'] = '2016-01-01'
params['output'] = 'csv'
```

```python
r = requests.get(url, params=params)
```

```python
df = pandas.read_csv(io.StringIO(r.content.decode('utf-8')))
```

```python
print(df)
```
